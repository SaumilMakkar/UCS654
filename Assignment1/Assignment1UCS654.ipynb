```python
import pandas as pd
import numpy as np
import warnings


warnings.filterwarnings('ignore')

def solve_pdf_assignment(roll_number, file_path='data.csv'):



    r = roll_number
    ar = 0.05 * (r % 7)
    br = 0.3 * ((r % 5) + 1)

    print(f"Roll Number: {r}")
    print(f"Calculated ar: {ar:.4f}")
    print(f"Calculated br: {br:.4f}")


    try:

        df = pd.read_csv(file_path, encoding='latin-1', usecols=lambda x: x.lower() == 'no2')
    except ValueError:

        try:
             df = pd.read_csv(file_path, encoding='latin-1', usecols=['NO2'])
        except:
             print("Error: Could not find 'no2' or 'NO2' column.")
             return


    df.columns = [c.lower() for c in df.columns]


    x = pd.to_numeric(df['no2'], errors='coerce').dropna().values


    x = x[x > 0]

    print(f"Valid records used: {len(x)}")


    z = x + ar * np.sin(br * x)



    mu_mle = np.mean(z)
    sigma2_mle = np.var(z)


    lambda_mle = 1 / (2 * sigma2_mle)


    c_mle = np.sqrt(lambda_mle / np.pi)



    print("FINAL PARAMETERS")

    print(f"λ (lambda): {lambda_mle:.6f}")
    print(f"μ (mu):     {mu_mle:.6f}")
    print(f"c:          {c_mle:.6f}")




    return lambda_mle, mu_mle, c_mle

if __name__ == "__main__":

    FILE_PATH = '/content/data[1].csv'
    ROLL_NUMBER = 102303862


    solve_pdf_assignment(ROLL_NUMBER, FILE_PATH)
```
