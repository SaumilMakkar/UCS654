```python




import threading
import os

def create_file(file_number):
    filename = f"file_{file_number}.txt"
    with open(filename, 'w') as f:
        f.write(f"This is file number {file_number}\n")
        f.write(f"Created by thread: {threading.current_thread().name}\n")
    print(f"Created: {filename}")

def main():
    os.makedirs("created_files", exist_ok=True)
    os.chdir("created_files")

    num_files = 10
    threads = []

    for i in range(1, num_files + 1):
        t = threading.Thread(target=create_file, args=(i,), name=f"Thread-{i}")
        threads.append(t)

    for t in threads:
        t.start()

    for t in threads:
        t.join()

    print(f"\nDone! {num_files} files created.")

if __name__ == "__main__":
    main()
```

```python
import threading
import os
import glob

def convert_to_uppercase(filepath):
    with open(filepath, 'r') as f:
        content = f.read()

    upper_content = content.upper()

    output_path = filepath.replace(".txt", "_upper.txt")
    with open(output_path, 'w') as f:
        f.write(upper_content)

    print(f"Converted: {os.path.basename(filepath)} → {os.path.basename(output_path)}")

def main():
    files = glob.glob("*.txt")

    if not files:
        print("No .txt files found! Run Mini-Project 1 first.")
        return

    threads = [threading.Thread(target=convert_to_uppercase, args=(f,)) for f in files]

    for t in threads:
        t.start()
    for t in threads:
        t.join()

    print(f"\nDone! {len(files)} files converted to uppercase.")

if __name__ == "__main__":
    main()
```

```python
import threading
import os
import glob
from PIL import Image   # pip install Pillow

# Since Cell 1 did os.chdir("created_files"), CWD is now inside created_files/
# Go one level up to get back to the Projects folder where the notebook lives
cwd = os.getcwd()
if os.path.basename(cwd) == "created_files":
    BASE_DIR = os.path.dirname(cwd)
else:
    BASE_DIR = cwd

IMAGES_DIR = os.path.join(BASE_DIR, "images")
GREY_DIR = os.path.join(BASE_DIR, "grey_images")

def convert_to_grey(image_path):
    img = Image.open(image_path)
    grey_img = img.convert("L")            # "L" = 8-bit greyscale

    filename = os.path.basename(image_path)
    name, ext = os.path.splitext(filename)
    output_path = os.path.join(GREY_DIR, f"{name}_grey{ext}")

    grey_img.save(output_path)
    print(f"Converted: {filename} → {os.path.basename(output_path)}")

def main():
    os.makedirs(GREY_DIR, exist_ok=True)

    images = (glob.glob(os.path.join(IMAGES_DIR, "*.jpg")) +
              glob.glob(os.path.join(IMAGES_DIR, "*.png")) +
              glob.glob(os.path.join(IMAGES_DIR, "*.bmp")))

    if not images:
        print(f"No images found! Add images to: {IMAGES_DIR}")
        return

    threads = [threading.Thread(target=convert_to_grey, args=(img,)) for img in images]

    for t in threads:
        t.start()
    for t in threads:
        t.join()

    print(f"\nDone! {len(images)} images converted to greyscale.")

if __name__ == "__main__":
    main()
```

```python
import threading
import os
import glob
import cv2    # pip install opencv-python

# CWD may be inside created_files/ due to Cell 1 — go up one level if needed
cwd = os.getcwd()
if os.path.basename(cwd) == "created_files":
    BASE_DIR = os.path.dirname(cwd)
else:
    BASE_DIR = cwd

VIDEOS_DIR = os.path.join(BASE_DIR, "videos")
GREY_VIDEOS_DIR = os.path.join(BASE_DIR, "grey_videos")

def convert_video_to_grey(video_path):
    cap = cv2.VideoCapture(video_path)

    fps    = cap.get(cv2.CAP_PROP_FPS)
    width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    filename = os.path.basename(video_path)
    name, ext = os.path.splitext(filename)
    output_path = os.path.join(GREY_VIDEOS_DIR, f"{name}_grey.avi")

    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height), isColor=False)

    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        grey_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        out.write(grey_frame)
        frame_count += 1

    cap.release()
    out.release()
    print(f"Converted: {filename} ({frame_count} frames) → {os.path.basename(output_path)}")

def main():
    os.makedirs(GREY_VIDEOS_DIR, exist_ok=True)

    videos = (glob.glob(os.path.join(VIDEOS_DIR, "*.mp4")) +
              glob.glob(os.path.join(VIDEOS_DIR, "*.avi")) +
              glob.glob(os.path.join(VIDEOS_DIR, "*.mov")))

    if not videos:
        print(f"No videos found! Add videos to: {VIDEOS_DIR}")
        return

    threads = [threading.Thread(target=convert_video_to_grey, args=(v,)) for v in videos]

    for t in threads:
        t.start()
    for t in threads:
        t.join()

    print(f"\nDone! {len(videos)} videos converted to greyscale.")

if __name__ == "__main__":
    main()
```
